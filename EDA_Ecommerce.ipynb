## Data Analysis

# DATA QUALITY CHECK

```python
print("\n=== DATA QUALITY ASSESSMENT ===")

#Check for missing values
print("\nMissing Values:\n")
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])

#Check for duplicates
print(f"\nDuplicate Rows: {df.duplicated().sum()}")
```

# Data Type Conversion

```python
# Convert date columns to datetime
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')
df['delivery_date'] = pd.to_datetime(df['delivery_date'], format='%d-%m-%Y')

#Check conversion
print("\nDate columns converted:\n")
print(f"Order Date: {df['order_date'].dtype}")
print(f"Delivery Date: {df['delivery_date'].dtype}")
```

# Data Cleaning

## i)Fill Null Values

```python
df['discount_%'] = df['discount_%'].fillna(0)
df['review_sentiment'] = df['review_sentiment'].fillna('Not Reviewed')

print("=== After Cleaning ===")
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])
```

## ii)remove duplicates

```python
df.drop_duplicates(inplace=True)
print("Rows after removing duplicates:", df.shape[0])
```

## iii) fix inconsistencies

```python
df['gender'] = df['gender'].str.strip().str.capitalize()
df['gender'] = df['gender'].replace({
    'M': 'Male',
    'F': 'Female',
})

df['payment_method'] = df['payment_method'].str.strip().str.title()
df['express_delivery'] = df['express_delivery'].str.strip().str.capitalize()
```

# UNIVARIATE ANALYSIS

```python
# Univariate Analysis is the simplest form of statistical analysis
# "uni" means "one" and "variate" means "variable"
# It is the analysis of a single variable

print("\n=== CATEGORICAL VARIABLES ===\n")

categorical_cols = ['gender', 'order_status', 'product_category',
                    'country', 'payment_method', 'shipping_carrier',
                    'express_delivery', 'review_sentiment',
                    'return_requested', 'customer_loyalty_tier']

for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())

    # Visualization
    plt.figure(figsize=(6, 6))
    df[col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

# NUMERICAL VARIABLE ANALYSIS

```python
print("\n=== NUMERICAL VARIABLES ===\n")

numerical_cols = ['age', 'quantity', 'unit_price',
                  'total_price', 'discount_%',
                  'delivery_days', 'rating']

for col in numerical_cols:
    print(f"\n--- {col} ---")
    print(f"Mean:    {df[col].mean():.2f}")
    print(f"Median:  {df[col].median():.2f}")
    print(f"Std Dev: {df[col].std():.2f}")
    print(f"Min:     {df[col].min():.2f}")
    print(f"Max:     {df[col].max():.2f}")

    # Visualization
    plt.figure(figsize=(8, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    df[col].hist(bins=30, color='lightgreen', alpha=0.7)
    plt.title(f'Distribution of {col}')

    # Boxplot
    plt.subplot(1, 2, 2)
    df.boxplot(column=col)
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

# BIVARIATE ANALYSIS

```python
# Bivariate Analysis is the statistical analysis of two variables at the same time.
# The core purpose is to investigate the relationship between them.

print("\n=== BIVARIATE ANALYSIS ===\n")

# ─────────────────────────────────────────
# 1. REVENUE & QUANTITY BY PRODUCT CATEGORY
# ─────────────────────────────────────────
print("\n=== REVENUE & QUANTITY BY PRODUCT CATEGORY ===\n")

category_performance = df.groupby('product_category').agg({
    'total_price': 'sum',
    'quantity'   : 'sum',
    'order_id'   : 'count'
}).rename(columns={'order_id': 'order_count'})

category_performance['avg_order_value'] = (
    category_performance['total_price'] /
    category_performance['order_count']
)

print(category_performance, '\n\n')

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Total Revenue by Category
category_performance['total_price'].plot(
    kind='bar', ax=axes[0,0], color='blue', alpha=0.7)
axes[0,0].set_title('Total Revenue by Category')
axes[0,0].tick_params(axis='x', rotation=30)

# Total Quantity by Category
category_performance['quantity'].plot(
    kind='bar', ax=axes[0,1], color='green', alpha=0.7)
axes[0,1].set_title('Total Quantity by Category')
axes[0,1].tick_params(axis='x', rotation=30)

# Avg Order Value by Category
category_performance['avg_order_value'].plot(
    kind='bar', ax=axes[1,0], color='orange', alpha=0.7)
axes[1,0].set_title('Avg Order Value by Category')
axes[1,0].tick_params(axis='x', rotation=30)

# Order Count by Category
category_performance['order_count'].plot(
    kind='bar', ax=axes[1,1], color='red', alpha=0.7)
axes[1,1].set_title('Order Count by Category')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────
# 2. SPENDING BY GENDER
# ─────────────────────────────────────────
print("\n=== SPENDING BY GENDER ===\n")

gender_performance = df.groupby('gender').agg({
    'total_price': 'sum',
    'quantity'   : 'sum',
    'order_id'   : 'count'
}).rename(columns={'order_id': 'order_count'})

gender_performance['avg_order_value'] = (
    gender_performance['total_price'] /
    gender_performance['order_count']
)

print(gender_performance, '\n\n')

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

gender_performance['total_price'].plot(
    kind='bar', ax=axes[0,0], color='blue', alpha=0.7)
axes[0,0].set_title('Total Revenue by Gender')
axes[0,0].tick_params(axis='x', rotation=30)

gender_performance['quantity'].plot(
    kind='bar', ax=axes[0,1], color='green', alpha=0.7)
axes[0,1].set_title('Total Quantity by Gender')
axes[0,1].tick_params(axis='x', rotation=30)

gender_performance['avg_order_value'].plot(
    kind='bar', ax=axes[1,0], color='orange', alpha=0.7)
axes[1,0].set_title('Avg Order Value by Gender')
axes[1,0].tick_params(axis='x', rotation=30)

gender_performance['order_count'].plot(
    kind='bar', ax=axes[1,1], color='red', alpha=0.7)
axes[1,1].set_title('Order Count by Gender')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────
# 3. REVENUE BY ORDER STATUS
# ─────────────────────────────────────────
print("\n=== REVENUE BY ORDER STATUS ===\n")

status_performance = df.groupby('order_status').agg({
    'total_price': 'sum',
    'quantity'   : 'sum',
    'order_id'   : 'count'
}).rename(columns={'order_id': 'order_count'})

status_performance['avg_order_value'] = (
    status_performance['total_price'] /
    status_performance['order_count']
)

print(status_performance, '\n\n')

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

status_performance['total_price'].plot(
    kind='bar', ax=axes[0,0], color='blue', alpha=0.7)
axes[0,0].set_title('Total Revenue by Order Status')
axes[0,0].tick_params(axis='x', rotation=30)

status_performance['quantity'].plot(
    kind='bar', ax=axes[0,1], color='green', alpha=0.7)
axes[0,1].set_title('Total Quantity by Order Status')
axes[0,1].tick_params(axis='x', rotation=30)

status_performance['avg_order_value'].plot(
    kind='bar', ax=axes[1,0], color='orange', alpha=0.7)
axes[1,0].set_title('Avg Order Value by Order Status')
axes[1,0].tick_params(axis='x', rotation=30)

status_performance['order_count'].plot(
    kind='bar', ax=axes[1,1], color='red', alpha=0.7)
axes[1,1].set_title('Order Count by Order Status')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()


# ─────────────────────────────────────────
# 4. REVENUE BY PAYMENT METHOD
# ─────────────────────────────────────────
print("\n=== REVENUE BY PAYMENT METHOD ===\n")

payment_performance = df.groupby('payment_method').agg({
    'total_price': 'sum',
    'quantity'   : 'sum',
    'order_id'   : 'count'
}).rename(columns={'order_id': 'order_count'})

payment_performance['avg_order_value'] = (
    payment_performance['total_price'] /
    payment_performance['order_count']
)

print(payment_performance, '\n\n')

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

payment_performance['total_price'].plot(
    kind='bar', ax=axes[0,0], color='blue', alpha=1)
axes[0,0].set_title('Total Revenue by Payment Method')
axes[0,0].tick_params(axis='x', rotation=30)

payment_performance['quantity'].plot(
    kind='bar', ax=axes[0,1], color='green', alpha=0.7)
axes[0,1].set_title('Total Quantity by Payment Method')
axes[0,1].tick_params(axis='x', rotation=30)

payment_performance['avg_order_value'].plot(
    kind='bar', ax=axes[1,0], color='orange', alpha=0.7)
axes[1,0].set_title('Avg Order Value by Payment Method')
axes[1,0].tick_params(axis='x', rotation=30)

payment_performance['order_count'].plot(
    kind='bar', ax=axes[1,1], color='red', alpha=0.7)
axes[1,1].set_title('Order Count by Payment Method')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

# Regional Performance

```python
print("\n=== COUNTRY PERFORMANCE ===\n")

country_performance = df.groupby('country').agg({
    'total_price': 'sum',
    'order_id':    'count'
}).rename(columns={'order_id': 'order_count'})

country_performance['avg_order_value'] = (
    country_performance['total_price'] / 
    country_performance['order_count']
)

print(country_performance, '\n\n')

# Visualization
plt.figure(figsize=(12, 8))

# Sales Distribution by Country
plt.subplot(2, 2, 1)
country_performance['total_price'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Country')

# Order Count by Country
plt.subplot(2, 2, 2)
country_performance['order_count'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Order Count by Country')

# Total Sales by Country bar chart
plt.subplot(2, 2, 3)
sns.barplot(x=country_performance.index, 
            y='total_price', 
            data=country_performance)
plt.title('Total Sales by Country')
plt.xticks(rotation=30)

# Average Order Value by Country
plt.subplot(2, 2, 4)
sns.barplot(x=country_performance.index, 
            y='avg_order_value', 
            data=country_performance)
plt.title('Average Order Value by Country')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

 # Time Series Analysis

```python
print("\n=== TIME SERIES ANALYSIS ===\n")

# Extract time components
df['order_month'] = df['order_date'].dt.to_period('M')
df['order_year']  = df['order_date'].dt.year

# Monthly trends
monthly_trends = df.groupby('order_month').agg({
    'total_price': 'sum',
    'order_id':    'count'
}).rename(columns={'order_id': 'order_count'})

print("Monthly Trends:")
print(monthly_trends.tail(10), '\n\n')

# Visualization
plt.figure(figsize=(15, 10))

# Monthly Sales Trend
plt.subplot(3, 1, 1)
monthly_trends['total_price'].plot(kind='line', 
                                    color='blue', 
                                    marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Total Sales')
plt.grid(True)

# Monthly Order Count Trend
plt.subplot(3, 1, 2)
monthly_trends['order_count'].plot(kind='line', 
                                    color='red', 
                                    marker='o')
plt.title('Monthly Order Count Trend')
plt.ylabel('Order Count')
plt.grid(True)

# Yearly Sales Comparison
plt.subplot(3, 1, 3)
yearly_trends = df.groupby('order_year')['total_price'].sum()
yearly_trends.plot(kind='bar', color='green', alpha=0.7)
plt.title('Yearly Sales Comparison')
plt.ylabel('Total Sales')
plt.xlabel('Year')
plt.xticks(rotation=0)
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Advanced Analysis

```python
print("\n=== Correlation Analysis ===\n")

#Select numerical columns for correlation

numerical_df = df[['age', 'quantity', 'unit_price',
                   'total_price', 'discount_%',
                   'delivery_days', 'rating']]

# Correlation matrix
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:")
print(correlation_matrix, '\n\n')

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,
            cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

# Customer Sales Analysis

```python
print("\n=== CUSTOMER LOYALTY ANALYSIS ===\n")

loyalty_analysis = df.groupby('customer_loyalty_tier').agg({
    'total_price': ['sum', 'mean'],
    'order_id':    'count',
    'discount_%':  'mean',
    'rating':      'mean'
}).round(2)

loyalty_analysis.columns = [
    'Total Sales', 'Avg Sales',
    'Order Count', 'Avg Discount', 'Avg Rating'
]

print(loyalty_analysis, '\n\n')

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Total Sales by Loyalty Tier
loyalty_analysis['Total Sales'].plot(
    kind='bar', ax=axes[0,0], color='lightblue')
axes[0,0].set_title('Total Sales by Loyalty Tier')
axes[0,0].tick_params(axis='x', rotation=30)

# Order Count by Loyalty Tier
loyalty_analysis['Order Count'].plot(
    kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('Order Count by Loyalty Tier')
axes[0,1].tick_params(axis='x', rotation=30)

# Average Rating by Loyalty Tier
loyalty_analysis['Avg Rating'].plot(
    kind='bar', ax=axes[1,0], color='orange')
axes[1,0].set_title('Average Rating by Loyalty Tier')
axes[1,0].tick_params(axis='x', rotation=30)

# Average Discount by Loyalty Tier
loyalty_analysis['Avg Discount'].plot(
    kind='bar', ax=axes[1,1], color='red')
axes[1,1].set_title('Average Discount by Loyalty Tier')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

# Key Insights Summary

```python
print("\n=== KEY INSIGHTS & SUMMARY ===\n")

# Overall Metrics
total_sales   = df['total_price'].sum()
total_orders  = df['order_id'].nunique()
avg_order_val = df['total_price'].mean()
avg_rating    = df['rating'].mean()

print("OVERALL PERFORMANCE:\n")
print(f"Total Sales:         ${total_sales:,.2f}")
print(f"Total Orders:        {total_orders}")
print(f"Avg Order Value:     ${avg_order_val:,.2f}")
print(f"Avg Customer Rating: {avg_rating:.2f}/5")

# Top Performing Categories
top_categories = df.groupby('product_category')['total_price']\
                   .sum()\
                   .sort_values(ascending=False)
print(f"\nTOP PERFORMING CATEGORIES:\n")
for i, (category, sales) in enumerate(
        top_categories.items(), 1):
    print(f"{i}. {category}: ${sales:,.2f}")

# Best Performing Country
best_country = df.groupby('country')['total_price']\
                 .sum()\
                 .idxmax()
best_country_sales = df.groupby('country')['total_price']\
                       .sum().max()
print(f"\nBEST PERFORMING COUNTRY: {best_country}")
print(f"Total Sales: ${best_country_sales:,.2f}")

# Delivery Performance
delivery_stats = df['order_status']\
                   .value_counts(normalize=True) * 100
print(f"\nDELIVERY STATUS:")
for status, percentage in delivery_stats.items():
    print(f"{status}: {percentage:.1f}%")

# Best Loyalty Tier
best_tier = df.groupby('customer_loyalty_tier')['total_price']\
              .sum()\
              .idxmax()
print(f"\nMOST VALUABLE LOYALTY TIER: {best_tier}")

# Payment Method Analysis
top_payment = df['payment_method']\
                .value_counts().head(3)
print(f"\nTOP 3 PAYMENT METHODS:")
for i, (method, count) in enumerate(
        top_payment.items(), 1):
    print(f"{i}. {method}: {count} orders")

# Shipping Carrier Analysis
top_carrier = df['shipping_carrier']\
                .value_counts().idxmax()
print(f"\nMOST USED SHIPPING CARRIER: {top_carrier}")

# Rating Analysis
print(f"\nRATING ANALYSIS:")
print(f"Average Rating:  {df['rating'].mean():.2f}")
print(f"Highest Rating:  {df['rating'].max():.1f}")
print(f"Lowest Rating:   {df['rating'].min():.1f}")
print(f"% Rated 4+:      "
      f"{(df['rating'] >= 4).sum() / df['rating'].notna().sum() * 100:.1f}%")

# Return Analysis
return_rate = (df['return_requested'] == 'Yes')\
                .sum() / len(df) * 100
print(f"\nRETURN ANALYSIS:")
print(f"Return Rate: {return_rate:.1f}%")

# Express Delivery Analysis
express_rate = (df['express_delivery'] == 'Yes')\
                 .sum() / len(df) * 100
print(f"\nEXPRESS DELIVERY RATE: {express_rate:.1f}%")

print("\n" + "="*50)
print("EDA COMPLETED SUCCESSFULLY!")
print("="*50)
```

# Key Insights & Summary

```python
print("\n=== EXPORT RESULTS ===\n")

# Collect all key metrics
analysis_results = {
    'total_sales':        total_sales,
    'total_orders':       total_orders,
    'avg_order_value':    avg_order_val,
    'avg_rating':         round(avg_rating, 2),
    'best_country':       best_country,
    'best_loyalty_tier':  best_tier,
    'top_category':       top_categories.index[0],
    'return_rate':        round(return_rate, 2),
    'express_delivery_rate': round(express_rate, 2),
}

# Convert to DataFrame and save
results_df = pd.DataFrame([analysis_results])
results_df.to_csv('eda_analysis_summary.csv', index=False)
print("Analysis summary saved to 'eda_analysis_summary.csv'")

# Also save cleaned dataset
df.to_csv('Ecommerce_orders_cleaned.csv', index=False)
print("Cleaned dataset saved to 'Ecommerce_orders_cleaned.csv'")

print("\n" + "="*50)
print("EDA PROJECT COMPLETED SUCCESSFULLY!")
print("="*50)
print("\nFiles saved:")
print("1. Ecommerce_orders.csv          -> Original dataset")
print("2. Ecommerce_orders_cleaned.csv  -> Cleaned dataset")
print("3. eda_analysis_summary.csv      -> Key metrics summary")
```
